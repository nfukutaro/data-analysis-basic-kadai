```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes =plt.subplots(2, 3, tight_layout=True)

#折れ線グラフを記載する
y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

#sin関数を記載する
#データを生成
# 0から10までの幅で配列を等間隔に10個作成する
x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)

axes[0,1].plot(x1,y1)
axes[0,1].set(xlim=(0,10), xticks=np.arange(0, 10),
              ylim=(-1,5), yticks=np.arange(-1,6))


#平均が15，標準偏差5の標準正規分布から2000個の乱数を生成する
x2 = np.random.normal(15, 5, 2000)

#ヒストグラムをプロット
axes[0,2].hist(x2,bins=10)
axes[0,2].set(xlim=(-5,40), xticks=np.arange(0,40,20),
              ylim=(0,600), yticks=np.arange(0,600,200))

#scatterのためのデータを生成する
np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2.5,len(x3))

#分散図を表す
axes[1,0].scatter(x3,y3)
axes[1,0].set(xlim=(-2,10), xticks=np.arange(0,10,5),
              ylim=(-5,10), yticks=np.arange(0,10,2.5))

plt.show()
```
